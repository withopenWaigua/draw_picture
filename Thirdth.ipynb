```python
N=100
df = pd.DataFrame(
    dict(group=np.repeat([1,2],N*2),
        y=np.append(
            np.append(np.random.normal(5,1,N),np.random.normal(2,1,N)),
            np.append(np.random.normal(1,1,N),np.random.normal(3,1,N))),
         x=np.tile(['A','B','A','B'],N)

        ) 
         )
```

```python
df
```

```python
# 1./fillgroup outlier=0='k'
# 2.  group+shape+alpha
(
    ggplot(df,aes(x='x',y='y',fill='factor(group)'))
    +geom_boxplot(outlier_size=0,color='k',position=position_dodge(0.85)) # 
    +geom_jitter(aes(group='factor(group)')
          ,shape='o'
          ,alpha=0.5
          ,position =position_jitterdodge(dodge_width = 0.85) # 
          
    )
    +theme_classic()
)
```
